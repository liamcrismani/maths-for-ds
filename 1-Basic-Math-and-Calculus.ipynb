## Functions


```python
def f(x: int|float) -> float:
    """Simple Cartesian function example"""
    return 2 * x + 1

x_values = [0, 1, 2, 3]

for x in x_values:
    y = f(x)
    print(y)
```

```python
# Charting a linear function in Python using SymPy
from sympy import *
x = symbols('x')
f = 2*x+1
plot(f)
```
